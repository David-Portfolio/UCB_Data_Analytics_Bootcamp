# Resample

```python
import numpy as np
import pandas as pd
from pathlib import Path
from collections import Counter
```

```python
# Load the data
df = pd.read_csv("..\\Starter_Code\\LoanStats_2019Q1.csv", skiprows=1)[:-2]
df
```

```python
# Drop the null columns where all values are null
df = df.dropna(axis="columns", how="any")
df = df.dropna(axis="rows", how="any")
df
```

```python
# Remove the `Issued` loan status
df = df[df.loan_status != "Issued"]
df = df[df.loan_status != "Fully Paid"]

# Add loan_risk1 column and assign values
df.insert(8, "loan_risk1", df.loan_status)
df.loan_risk1.replace(["Current", "Charged Off", "In Grace Period", "Late (16-30 days)", "Late (31-120 days)"], 
                      ["0", "1", "1", "1", "1"], inplace=True)

df
```

```python
# Assign loan_risk1 as target column and to numeric
target_column = ["loan_risk1"]
df["loan_risk1"] = pd.to_numeric(df["loan_risk1"])
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
df = df.select_dtypes(include=numerics)
df
```

```python
df.dtypes
```

```python
df.to_csv("..\\Starter_Code\\LoanStats_2019Q1_updated.csv")
```

```python
# Define features set with high_risk
X = df.copy()
X = X.drop("loan_risk1", axis=1)
X
```

```python
# Define target vector with low_risk
y = df["loan_risk1"].values
y
```

```python
X.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Create X_train, X_test, y_train, y_test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
# Create StandardScaler instance
scaler = StandardScaler()
```

```python
# Fit Standard Scaler
X_scaler = scaler.fit(X_train)
```

```python
# Scale data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Oversample

### Random Oversample

```python
from imblearn.over_sampling import RandomOverSampler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import balanced_accuracy_score
from imblearn.metrics import classification_report_imbalanced
```

```python
# Random Oversampling
random_oversampler = RandomOverSampler(random_state=1)
X_resampled, y_resampled = random_oversampler.fit_resample(X_train, y_train)

Counter(y_resampled)
```

```python
# Train Logistic Regression model using Random Oversampled data
model = LogisticRegression(solver='lbfgs', random_state=1)
model.fit(X_resampled, y_resampled)
```

```python
# Display the confusion matrix
y_pred = model.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
# Calculate the balanced accuracy score
balanced_accuracy_score(y_test, y_pred)
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y_test, y_pred))
```

### SMOTE Oversample

```python
from imblearn.over_sampling import SMOTE
```

```python
# Resample the training data with SMOTE
X_resampled, y_resampled = SMOTE(random_state=1, sampling_strategy=1.0).fit_resample(X_train, y_train)
Counter(y_resampled)
```

```python
# Train the Logistic Regression model using the resampled data
model = LogisticRegression(solver='lbfgs', random_state=1)
model.fit(X_resampled, y_resampled)
```

```python
# Calculated the balanced accuracy score
y_pred = model.predict(X_test)
balanced_accuracy_score(y_test, y_pred)
```

```python
# Display the confusion matrix
confusion_matrix(y_test, y_pred)
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y_test, y_pred))
```

# Undersample

```python
from imblearn.under_sampling import ClusterCentroids
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import balanced_accuracy_score
from imblearn.metrics import classification_report_imbalanced
```

```python
# Resample the data using the ClusterCentroids resampler
cluster = ClusterCentroids(random_state=1)
X_resampled, y_resampled = cluster.fit_resample(X_train, y_train)
Counter(y_resampled)
```

```python
# Logistic regression using cluster centroid undersampled data
model = LogisticRegression(solver='lbfgs', random_state=1)
model.fit(X_resampled, y_resampled)
```

```python
# Display the confusion matrix
y_pred = model.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
# Calculate the balanced accuracy score
balanced_accuracy_score(y_test, y_pred)
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y_test, y_pred))
```
